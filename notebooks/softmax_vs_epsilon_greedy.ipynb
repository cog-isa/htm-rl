```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
x = np.arange(10)
```

Изменение температуры softmax

```python
for beta in np.linspace(0, 1, 10):
    y = softmax(beta*x)
    plt.plot(x, y, label=f'{round(beta, 1)}')
plt.legend()
plt.grid()
```

Масштабируется разница мажду вариантами, выделенное нарпавление сохраняется, но размывается

Добавление epsilon шума.

```python
for beta in np.linspace(0, 1, 10):
    y = softmax(x)
    y = (1 - beta) / x.size + beta * y
    plt.plot(x, y, label=f'{round(beta, 1)}')
plt.legend()
plt.grid()
```

Разница между вариантами тоже масштабируется, но более равномерно.

```python
beta = 0.7
(1 - beta) / x.size + beta * y
```
